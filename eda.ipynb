```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset into a pandas dataframe
df = pd.read_csv("F:\py\ml\heart-disease-prediction\heart_failure_clinical_records_dataset.csv")

# Print the first 5 rows of the dataset
print(df.head())

# Print some basic information about the dataset
print(df.info())

# Print summary statistics for each column
print(df.describe())

# Plot a histogram of the age distribution by DEATH_EVENT
sns.histplot(data=df, x="age", hue="condition", multiple="stack")
plt.show()

# Plot a scatterplot of serum creatinine vs. ejection fraction by DEATH_EVENT
sns.scatterplot(data=df, x="ejection_fraction", y="serum_creatinine", hue="condition")
plt.show()

# Plot a boxplot of the age distribution by smoking status and DEATH_EVENT
sns.boxplot(data=df, x="smoking", y="age", hue="condition")
plt.show()

```
