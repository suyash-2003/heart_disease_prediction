```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset into a pandas dataframe
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')

# Print the first 5 rows of the dataset
print(df.head())

# Print some basic information about the dataset
print(df.info())

# Print summary statistics for each column
print(df.describe())

# Plot a histogram of the age distribution by DEATH_EVENT
sns.histplot(data=df, x="age", hue="DEATH_EVENT", multiple="stack")
plt.show()

# Plot a scatterplot of serum creatinine vs. ejection fraction by DEATH_EVENT
sns.scatterplot(data=df, x="ejection_fraction", y="serum_creatinine", hue="DEATH_EVENT")
plt.show()

# Plot a boxplot of the age distribution by smoking status and DEATH_EVENT
sns.boxplot(data=df, x="smoking", y="age", hue="DEATH_EVENT")
plt.show()

# Plot a heatmap of the correlation matrix by DEATH_EVENT
sns.heatmap(df[df.columns.difference(['DEATH_EVENT'])].corrwith(df['DEATH_EVENT']), cmap="coolwarm", annot=True)
plt.show()
```
